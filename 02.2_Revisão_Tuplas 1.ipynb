# Tuplas

Tuplas podem conter todos os tipos de dados:

image.png

````python
# Criando uma tupla com os três tipos de dados acima

tupla_a = (1, 4.0, "Centro")
tupla_a
````

````python
# Verificando o tipo da variável

type(tupla_a)
````

## Indexação

Os elementos de uma tupla podem ser acessados por seus índices. A tupla funciona como uma espécie de tabela ordenando os elementos pelo ID.

A forma de chamar o objeto específico em uma tupla é:

```
nome_da_tupla[ID]
```
Por exemplo:
```python
tupla_a[0]
```

image.png

````python
# Imprimindo os valores armazenados segundo seus índices (IDs)

print(tupla_a[0])
print(tupla_a[1])
print(tupla_a[2])
````

````python
# Verificando o tipo de cada valor armazenado na tupla

print(type(tupla_a[0]))
print(type(tupla_a[1]))
print(type(tupla_a[2]))
````

Também podemos utilizar índices negativos

image.png

````python
# Imprimindo os valores armazenados segundo seus índices (IDs) negativos

print(tupla_a[-1])
print(tupla_a[-2])
print(tupla_a[-3])
````

## Concatenação

````python
# Para concatenar tuplas basta utilizar o sinal de '+'

tupla_b = tupla_a + ("Treinamento", 10)
tupla_b
````

## Fatiamento

````python
# Fatiando tuplas

tupla_b[2:5]
````

````python
# Do início a um ID

tupla_b[:4]
````

````python
# De algum ID ao final da tupla

tupla_b[3:]
````

````python
# Verificando o comprimento da tupla com a função 'len()' (length)

len(tupla_b)
````

## Ordenação

Podemos classificar os valores de uma tupla e salvá-los em uma nova tupla:

````python
# Uma tupla de exemplo

valores = (0, 9, 6, 5, 10, 8, 9, 6, 2)
````

````python
# Ordene a tupla

valores_ordenados = sorted(valores)
valores_ordenados
````

## Aninhamento

Podemos ainda "aninhar" tuplas, listas e mesmo dicionários, que significa armazenar objetos de coleção dentro de uma tupla.

````python
tup = (1, 2, {"chave a": 5, "chave b": 6})
tup
````

````python
# Criando tupla com outras aninhadas
tuplas_aninhadas =(4.0, "Centro", (2, True), ("Treinamento", 10), ("Desenvolvimento", (False, 42)))
tuplas_aninhadas
````

````python
# Imprimindo os elementos por seus IDs

print("Elemento 0 da Tupla: ", tuplas_aninhadas[0])
print("Elemento 1 da Tupla: ", tuplas_aninhadas[1])
print("Elemento 2 da Tupla: ", tuplas_aninhadas[2])
print("Elemento 3 da Tupla: ", tuplas_aninhadas[3])
print("Elemento 4 da Tupla: ", tuplas_aninhadas[4])
````

````python
# Escreva os elementos em cada índice, considerando os aninhados

print("Elemento 0 da Tupla:\t", tuplas_aninhadas[0])
print("Elemento 1 da Tupla:\t", tuplas_aninhadas[1])
print("Elemento 2, 0 da Tupla:\t", tuplas_aninhadas[2][0])
print("Elemento 2, 1 da Tupla:\t", tuplas_aninhadas[2][1])
print("Elemento 3, 0 da Tupla:\t", tuplas_aninhadas[3][0])
print("Elemento 3, 1 da Tupla:\t", tuplas_aninhadas[3][1])
print("Elemento 4, 0 da Tupla:\t", tuplas_aninhadas[4][0])
print("Elemento 4, 1, 0 da Tupla:\t", tuplas_aninhadas[4][1][0])
print("Elemento 4, 1, 1 da Tupla:\t", tuplas_aninhadas[4][1][1])
````

Podemos acessar strings nas tuplas aninhadas usando um índice a mais:

````python
# Abreviando Treinamento

tuplas_aninhadas[3][0][:7]
````

````python
# Abreviando Desenvolvimento

tuplas_aninhadas[4][0][:6]
````
