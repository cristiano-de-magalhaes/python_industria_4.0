```python
# criei essa função para "abrir" todos os ramos dos nós
def print_list(obj, nivel=0):
    lst = obj.get_children()
    print('    ' * (nivel - 1) + '+-- ' * (nivel > 0) + str(lst[0]))
    for l in lst[:]:
        if type(l) is list:
            print_list(l, nivel + 1)
        elif type(l) is tipo and len(l.get_children()) > 0:
            print_list(l, nivel + 1)
        else:
            print('    ' * nivel + '+-- ' + str(l))

print_list(root)
```
