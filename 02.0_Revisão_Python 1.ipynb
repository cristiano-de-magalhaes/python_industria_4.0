# Conversão entre tipos diferentes em Python 3.x

A ação de mudar o tipo de um determinado objeto é chamado de `typecasting`.

## Entre inteiros e pontos flutuantes


```python
# Converte 42 para float
a = int(42)       # Cria um objeto inteiro 'a'
b = float(a)      # Converte o objeto inteiro 'a' em um float 'b'
print(b, type(b)) # Imprime o valor de 'b' e seu tipo (type)
```

Ao se converter um `int` para um `float`, não há perda de informação pois toda informação contida no `int` pode ser representada por um `float`.

Já no caminho inverso há perda de informação referente à porção fracional ou decimal do valor.


```python
# Converte 42.6 para inteiro
a = float(42.6)   # Cria um objeto float 'a'
b = int(a)        # Converte o objeto float 'a' em um int 'b'
print(b, type(b)) # Imprime o valor de 'b' e seu tipo (type)
```

## Conversões entre Strings, inteiros e pontos flutuantes

Caso um objeto string contenha um valor numérico, ela pode ser convertida para um inteiro ou ponto flutuante.

```python
# Converte uma string em um inteiro

int('12')
```

Mas isso só funciona caso a string contenha SOMENTE valores numéricos.

```python
# A conversão abaixo retorna erro

int('3 a 4')
```

O mesmo vale para representação de números em formato de ponto flutuante:


```python
# Converte a string "42.6" em um float

float('42.6')
```

A conversão inversa também é possível:

```python
# Converte um int em uma string

str(42)
```

```python
# Converte um float em uma string

str(42.6)
```

## Convertendo valores booleanos

Também é possível a conversão de valores booleanos.

Valor `bool` | Valor `int` | Valor `float`
:-:|:-:|:-:
True|1|1.0
False|0|0.0

```python
# Convertendo True para int

int(True)
```

```python
# Convertendo 1 para bool

bool(1)
```

```python
# Convertendo 0 para booleano

bool(0)
```

```python
# Convertendo True para float

float(True)
```

## Variáveis

Podemos armazenar valores em variáveis para utilização posterior:

```python
# Armazena valor na variável 'x'

x = 42
```

```python
# Exibe o valor armazenado na variável 'x'

x
```

```python
# Armazenando o resultado da operação em outra variável

y = x / 6
y
```

```python
# Atualizando valor na própria variável

x = x / 7
x
```

```python
# Agrupando operações para permitir complexidade

total = (42 * 2) + (3 ** 4) + (90 // 4) / (25 % 6)
total
```
